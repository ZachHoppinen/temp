```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('./data/lab_test_2.dat', skiprows = 1)
units = df.iloc[0]
stat_meth = df.iloc[1]
data = df[2:]
data = data.set_index('TIMESTAMP')

# Converting the index as date
data.index = pd.to_datetime(data.index)

#converting to type numeric
data = data.replace('NAN',np.nan)
data = data.dropna(axis =1)
data = data.apply(pd.to_numeric)
```

```python
data
```

```python
fig = px.line(data, x=data.index, y=data.columns.drop(['NR01TK_Avg','RECORD','Albedo_Avg']),
              title='Lab Test Data',labels={'value':'Radiation (W/m^2)','TIMESTAMP':''})
fig.show()
```

```python
fig.write_html('./figures/nr01_20211018.html')
```
