```python
import pandas as pd
```

```python
df = pd.read_csv('../../banner/temp_array/data/raw/summer2020/CR1000_Banner2020_FiveMinute.dat', skiprows = 1)
```

```python
df = df[2:]
```

```python
df.loc[:,'datetime'] = pd.to_datetime(df.TIMESTAMP)
df = df.drop('TIMESTAMP',axis=1)
df = df.set_index('datetime')
df = df.sort_index()
```

```python
for col in df.columns:
    try:
        df[col] = df[col].astype('f8')
    except:
        pass
```
